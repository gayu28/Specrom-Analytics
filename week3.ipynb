```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib
import requests
import urllib
from fake_useragent import UserAgent
import requests
from lxml.html import fromstring
import requests
from itertools import cycle
import traceback
import requests
from bs4 import BeautifulSoup
from random import choice
from fake_useragent import UserAgent
import random
import time
```

```python
def apply1(input):
    query = input
    address = "http://www.bing.com/search?q=%s" % (
        urllib.parse.quote_plus(query))  
    ua = UserAgent()
    headers = {'User-Agent': ua.random}
    
    def proxy_generator():
        response = requests.get("https://free-proxy-list.net/")
        soup = BeautifulSoup(response.content, 'html5lib')
        proxy = {'https': choice(list(map(lambda x: x[0] + ':' + x[1], list(
            zip(map(lambda x: x.text, soup.findAll('td')[::8]), map(lambda x: x.text, soup.findAll('td')[1::8]))))))}
        return proxy
    
    proxy = proxy_generator()
    print("Proxy currently being used: {}".format(proxy))
    print("current user agent: {} ".format(ua.random))
    for attempt in range(1, 9):

        try:

            seconds = random.randint(1, 3)
            r = requests.get(address, time.sleep(seconds), headers=headers, proxies=proxy, timeout=7)

            # print(r.status_code)
            html_response = r.text
            return_dict = {}
            # print(return_dict)

            return_dict["status_code"] = r.status_code

            if return_dict["status_code"] == 200:
                soup = BeautifulSoup(html_response, 'html.parser')
                [s.extract() for s in soup('span')]
                results = soup.findAll('li', {"class": "b_algo"})
                temp_list = []
                i = 1
                for result in results:
                    temp_dict = {}
                    intermediate_result = result.find('h2')
                    temp_dict["position"] = str(i)
                    try:
                        temp_dict["url"] = intermediate_result.find('a')['href']
                    except Exception as E:
                        temp_dict["url"] = ''

                    try:
                        temp_dict["title"] = intermediate_result.get_text()
                    except Exception as E:
                        temp_dict["title"] = ''

                    try:
                        temp_dict["snippet"] = result.find('p').get_text()
                    except Exception as E:
                        temp_dict["snippet"] = ''

                    temp_list.append(temp_dict)
                    i = i + 1
                return_dict["Results"] = temp_list
                return return_dict
        except:
            print(attempt)
            print("Connection error, Query_string:", input)
            return_dict["Results"] = "Unable to fetech the query " + input + " in 5 attempts"
            continue
        else:
            print("finished request in attempt ", attempt)

    return return_dict
```

##### Generation of output CSV
- apply1() function is called here and output dataframe is generated


```python
df=pd.read_csv('gayathri week 3 assignment.csv')temp_li=[]
for i in df['Name (en-GB)']:
    temp_dict=apply1(i)
    temp_df=pd.DataFrame.from_dict(temp_dict['Results'])
    temp_df['Query String']=i
    temp_li.append(temp_df)
    
out_df=pd.concat(temp_li,ignore_index=True)    
```

```python
out_df=pd.concat(temp_li,ignore_index=True, sort='True')   
out_df.to_csv('partial_op.csv')
```
